```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader('/Users/hwangjunho/Desktop/encore_skn11/08_LLM/04_langchain/data/The_Adventures_of_Tom_Sawyer.pdf')
documents = loader.load()
print(len(documents))
print(documents[0].metadata)
print('=' * 100)
print(documents[2].page_content)
```

```python
### Embedding Model
```

```python
from langchain_openai import OpenAIEmbeddings

embedding_model = OpenAIEmbeddings(model="text-embedding-3-small")

text = "The quick brown fox jumps over the lazy dog"
embedding_model.embed_query(text)
```

```python
doc = [document.page_content for document in documents]
vects = embedding_model.embed_documents(doc)

print(len(vects))
```

### Vector Store

```python
from langchain.vectorstores import FAISS

txt_embeddings = list(zip(doc, vects))
vector_store = FAISS.from_embeddings(txt_embeddings, embedding_model)
vector_store = FAISS.from_documents(documents, embedding_model)
```

```python
vector_store.similarity_search("Tom Sawyer", k=3)
```

### Retriever

```python
from langchain.chains import RetrievalQA
from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    model_name='gpt-4o-mini',
    temperature=0
)

retriever = vector_store.as_retriever()

retrieval_qa = RetrievalQA.from_chain_type(
    llm=model,
    retriever=retriever,
    chain_type='stuff'
)
```

```python
response = retrieval_qa.invoke('"마을 무덤에 있던 남자는 누가 죽였나요?')
response
```
