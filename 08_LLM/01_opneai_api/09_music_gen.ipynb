# 멜로디 생성하기

```python
from transformers import pipeline
import scipy

synthesier = pipeline("text-to-audio", "facebook/musicgen-small")

music = synthesier("festival music with a spring melody", forward_params={"do_sample": True})

scipy.io.wavfile.write("musicgen_output.wav", rate=music["sampling_rate"], data=music["audio"])
```
