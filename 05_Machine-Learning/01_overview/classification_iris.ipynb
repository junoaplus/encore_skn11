```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.font_manager as fm
import matplotlib
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier

font_path = '/Users/hwangjunho/Desktop/encore_skn11/04_data-analysis/03_matplotlib/AppleSDGothicNeo.ttc'
font = fm.FontProperties(fname=font_path).get_name()
matplotlib.rc('font', family=font)
```

```python
iris_data = load_iris()
iris_data.keys()
iris_data.target_names
iris_data.feature_names 
```

```python
print(iris_data.data)   # input 데이터 (특성 데이터)
print(iris_data.feature_names)
len(iris_data.data)
```

```python
print(iris_data.target)     # 숫자로 인코딩된 라벨 데이터 (정답 데이터)
print(iris_data.target_names)   # 라벨 데이터의 이름
len(iris_data.target)
```

```python
# 데이터 전처리
# 1. 입력-라벨 데이터
# 2. 학습용-테스트용 데이터
# 3. 데이터 정규화
iris_input = iris_data.data
iris_label = iris_data.target
train_input, test_input, train_label, test_label = train_test_split(iris_input, iris_label, test_size=0.3, stratify=iris_label, random_state=10)

train_input, test_input, train_label, test_label
print(len(train_input), len(test_input))
print(len(train_label), len(test_label))
 
mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)
print(mean, std)

train_scaled = (train_input - mean) / std
test_scaled= (test_input - mean) / std
print(train_scaled, test_scaled)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(train_input)
train_scaled = scaler.transform(train_input)
test_scaled = scaler.transform(test_input)
```

```python
# 훈현 
kn = KNeighborsClassifier()
kn.fit(train_scaled, train_label)
```

```python
# 평가 score
kn.score(test_scaled, test_label)
```

```python
# 평가 : accuracy_score
from sklearn.metrics import accuracy_score
pred = kn.predict(test_scaled)
accuracy_score(test_label, pred)
```
